````python
import os
import moviepy.video.io.ImageSequenceClip
````

### TODO

````python
# fill in the path on your machine
vid_name = 'vid_14' # name of the video to process
box_frames_folder = "/data/huanga/Synchrony/box_frames" # folder containing frames generated by videos_to_frames.ipynb
box_video_folder = "/data/huanga/Synchrony/box_videos" # folder to save the video
os.makedirs(box_video_folder, exist_ok=True)

fps = 30.0 # frames per second
````

### Compile to video

NOTE: Sometimes you might need to re-run  the following command for moviepy to work. See <a href="https://stackoverflow.com/questions/68032884 getting-typeerror-must-be-real-number-not-nonetype-whenever-trying-to-run-wr"> this thread </a> for more information.

````python
image_files = [os.path.join(box_frames_folder, vid_name, img)
               for img in sorted(os.listdir(os.path.join(box_frames_folder, vid_name)))
               if img.endswith(".jpg")]
clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(image_files, fps=fps)
clip.write_videofile(os.path.join(box_video_folder, vid_name + ".mp4"))
````
